````python
# Import necessary libraries
import pandas as pd

# Load the World Bank dataset directly from their official data repository
url = 'https://api.worldbank.org/v2/en/indicator/AG.LND.FRST.ZS?downloadformat=csv'

# Download and unzip data automatically
import requests, zipfile, io
r = requests.get(url)
z = zipfile.ZipFile(io.BytesIO(r.content))
z.extractall("forest_data")


````

````python
# Load data from the extracted CSV (the correct file typically contains 'API_AG.LND.FRST.ZS' in filename)
df = pd.read_csv('forest_data/API_AG.LND.FRST.ZS_DS2_en_csv_v2_19393.csv', skiprows=4)

# Transform data from wide to long format
df_long = pd.melt(df,
                  id_vars=['Country Name', 'Country Code'],
                  value_vars=[str(year) for year in range(1990, 2022)],
                  var_name='Year',
                  value_name='ForestPercent')

# Drop rows with missing values
df_long.dropna(subset=['ForestPercent'], inplace=True)

# Save as CSV
df_long.to_csv('forest_area_long.csv', index=False)
````

````python
# Notebook: Day 4 – Biodiversity & Deforestation Mapping
import pandas as pd
import plotly.express as px

# Load FAO Forest Area dataset (long format)
df_forest = pd.read_csv('forest_area_long.csv')

# Choropleth animation
fig = px.choropleth(df_forest, locations="Country Code",
                    color="ForestPercent", hover_name="Country Name",
                    animation_frame="Year", color_continuous_scale='Greens',
                    range_color=[0, 100],
                    title='Global Forest Cover Change (1990–2020)')
fig.show()

# --- Starter Notebook elide: remove animation_frame argument ---
````
